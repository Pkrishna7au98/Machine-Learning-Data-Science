```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
```

```python
data.data
```

```python
data.feature_names
```

```python
data.target
```

```python
data.target_names
```

```python
#Create DataFrame

df = pd.DataFrame(np.c_[data.data , data.target], columns = [list(data.feature_names)+['target']])
df.head()
```

```python
df.shape
```

##### SPLIT DATA

```python
x= df.iloc[:,0:-1]
y= df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2 , random_state =22)

print('shape of x_train = ', x_train.shape)
print('shape of x_test = ', x_test.shape)
print('shape of y_train = ', y_train.shape)
print('shape of y_test = ', y_test.shape)
```

### Train Decision Tree Classification Model

```python
from sklearn.tree  import DecisionTreeClassifier
```

```python
classifier_gini = DecisionTreeClassifier(criterion = 'gini')

classifier_gini.fit(x_train , y_train)
```

```python
classifier_gini.score(x_test, y_test)
```

Wow a great accuracy.

```python
classifier_entr = DecisionTreeClassifier(criterion = 'entropy')
classifier_entr.fit(x_train , y_train)
classifier_entr.score(x_test, y_test)
```

The accuracy is less than GINI , so we use the above one
