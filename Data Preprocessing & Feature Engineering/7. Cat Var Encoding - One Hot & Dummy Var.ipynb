## Dummy Variable and one Hot Encoding

```python
import pandas as pd
```

```python
tips_df = pd.read_csv(r"C:\Users\KRISHNA\Downloads\Compressed\house-prices-advanced-regression-techniques\DATASETS\seaborn-data\tips.csv")
tips_df
```

```python
dummy = pd.get_dummies(tips_df)     #We will get new columns made up of each column having categorical variable but in form of 0 and 1.      
dummy                              # For example, Sex column has two nominal categorical vars, so we got two new columns
                                    # sex_male and sex_female .  # For K dummy variables
```

```python
dummy_df = pd.get_dummies(tips_df , drop_first=True)
dummy_df                                              # For K-1 dummy variables
```

#### Now coverting the categorical data to One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oh_enc = OneHotEncoder(sparse = False)  # For K dummy variables
```

```python
oh_enc_arr = oh_enc.fit_transform(tips_df[['sex','smoker','day','time']])
oh_enc_arr
```

```python
dummy.keys()
```

```python
oh_enc_df = pd.DataFrame(oh_enc_arr, columns=['sex_Female', 'sex_Male', 'smoker_No',
       'smoker_Yes', 'day_Fri', 'day_Sat', 'day_Sun', 'day_Thur',
       'time_Dinner', 'time_Lunch'])
oh_enc_df  
```

```python
oh_enc = OneHotEncoder(sparse = False , drop = 'first')  # For K-1 dummy variables
```

```python
oh_enc_arr2 = oh_enc.fit_transform(tips_df[['sex','smoker','day','time']])
oh_enc_arr2
```
